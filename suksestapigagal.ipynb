```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
ndelok = pd.read_csv('/content/drive/MyDrive/Latihan/ini.csv')
ndelok
```

```python
ndelok_c = ndelok.astype('category')

ndelok_c["Posisi Sayap"] = ndelok_c["Posisi Sayap"].cat.codes
ndelok_c["Kemiringan Sayap"] = ndelok_c["Kemiringan Sayap"].cat.codes
ndelok_c["Bentuk Sayap"] = ndelok_c["Bentuk Sayap"].cat.codes
ndelok_c["Arah Sayap"] = ndelok_c["Arah Sayap"].cat.codes
ndelok_c["Jenis Mesin"] = ndelok_c["Jenis Mesin"].cat.codes
ndelok_c["Jumlah Mesin"] = ndelok_c["Jumlah Mesin"].cat.codes
ndelok_c["Posisi Mesin"] = ndelok_c["Posisi Mesin"].cat.codes
ndelok_c["Bentuk Badan"] = ndelok_c["Bentuk Badan"].cat.codes
ndelok_c["Hidung Badan"] = ndelok_c["Hidung Badan"].cat.codes
ndelok_c["Tengah Badan"] = ndelok_c["Tengah Badan"].cat.codes
ndelok_c["Posisi Ekor"] = ndelok_c["Posisi Ekor"].cat.codes
ndelok_c["Jumlah Ekor"] = ndelok_c["Jumlah Ekor"].cat.codes
ndelok_c["Bentuk Ekor"] = ndelok_c["Bentuk Ekor"].cat.codes
ndelok_c["Persenjataan"] = ndelok_c["Persenjataan"].cat.codes
ndelok_c["Warna"] = ndelok_c["Warna"].cat.codes
ndelok_c["Nama PTTA"] = ndelok_c["Nama PTTA"].cat.codes

ndelok_c.head(20)
```

```python
X = ndelok_c.iloc[:,:15].values
Y = ndelok_c.iloc[:, 15].values
```

```python
Y = ndelok_c.iloc[:, 15].values
Y
```

```python
#test size = 0.20, which means 80% = training and 20% testing, random state = 45
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1 , random_state=45)
print(X_train.shape, y_train.shape)
```

```python
#create a Gaussian classifier
model = GaussianNB()

#train models with training data
model.fit(X_train,y_train)

#response prediction for test data
y_pred = model.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```
